```python
# Importing libraries necessary for this assignment
import pandas as pd # Pandas for data manipulation
# Matplotlib and Seaborn for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np # Numpy for mathematical calculations
import statsmodels.formula.api as smf # regression modeling
from sklearn.model_selection import train_test_split # Train_test_split for splitting dataset into training and testing
import sklearn.linear_model # linear modeling in scikit-learn

# Setting pandas display options 
# to ensure better visibility of rows and columns
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)


# specifying file name
file = './datasets/Apprentice_Chef_Dataset_2023.xlsx'


# reading the file into Python
customer = pd.read_excel(io = file)

#changing mislabeled column name
customer.rename(columns={'LARGEST_ORDER_SIZE': 'AVG_MEAL_ORDER_PER_CUSTOMER'},inplace=True)
customer.rename(columns={'LATE_DELIVERIES ': 'LATE_DELIVERIES'},inplace=True)

# outputting the first ten rows of the dataset
customer.head(n = 10)
```

```python
# Checking for missing values in the dataset and summing them up
customer.isnull().sum()
```

```python
# Filling missing values in 'FAMILY_NAME' column with 'N/A'
customer['FAMILY_NAME'] = customer['FAMILY_NAME'].fillna('N/A')
```

```python
# Creation of new columns (feature enginering)

# Average Delivery Time
# Calculate average delivery time for each customer
customer['AVG_DELIVERY_TIME'] = customer['LATE_DELIVERIES'] / customer['TOTAL_MEALS_ORDERED'].round(decimals = 2)

# Total login
# Calculate the total number of logins (desktop and mobile) for each customer
customer['TOTAL_LOGIN'] = customer['PC_LOGINS'] + customer['MOBILE_LOGINS']

# Desktop Login Ratio
#Calculate the ratio of desktop logins to total logins for each customer
customer['DESKTOP_LOGIN_RATIO'] = customer['PC_LOGINS'] / customer['TOTAL_LOGIN'].round(decimals = 2)

# Mobile Login Ratio
# Calculate the ratio of mobile logins to total logins for each customer
customer['MOBILE_LOGIN_RATIO'] = customer['MOBILE_LOGINS'] / customer['TOTAL_LOGIN'].round(decimals = 2)

# Average Order Value
# Calculate the average order value for each customer
customer['AVG_ORDER_VALUE'] = customer['REVENUE'] / customer['TOTAL_MEALS_ORDERED'].round(decimals = 2)

# Promotion Participation
# Identify if the customer participates in promotions (1 if yes, 0 if no)
customer['PROMO_PARTICIPATION'] = (customer['WEEKLY_PLAN'] > 0).astype(int)

# Customer Engagement Score
# Calculate the customer engagement score based on multiple factors
customer['CUSTOMER_ENGAGEMENT_SCORE'] = customer['CONTACTS_W_CUSTOMER_SERVICE'] + \
                                        customer['PRODUCT_CATEGORIES_VIEWED'] + \
                                        customer['TOTAL_PHOTOS_VIEWED'] + \
                                        customer['TOTAL_MEALS_ORDERED'] + \
                                        customer['AVG_TIME_PER_SITE_VISIT'].round(decimals = 2)

# Meal Plan Utilization
# Calculate the ratio of weekly plan to total meals ordered
customer['MEAL_PLAN_UTILIZATION'] = customer['WEEKLY_PLAN'] / customer['TOTAL_MEALS_ORDERED'].round(decimals = 2)

# Cancel Rate
# Calculate the cancel rate (cancellations after noon per total meals ordered)
customer['CANCEL_RATE'] = customer['CANCELLATIONS_AFTER_NOON'] / customer['TOTAL_MEALS_ORDERED'].round(decimals = 2)
```

```python
# building a full model

# blueprinting a model type
lm_full = smf.ols(formula = """  REVENUE ~ UNIQUE_MEALS_PURCH+ PRODUCT_CATEGORIES_VIEWED+ LATE_DELIVERIES+ TOTAL_LOGIN+  
                                TOTAL_PHOTOS_VIEWED+ AVG_TIME_PER_SITE_VISIT+ AVG_DELIVERY_TIME+ AVG_ORDER_VALUE+
                                 CUSTOMER_ENGAGEMENT_SCORE+ AVG_MEAL_ORDER_PER_CUSTOMER+ MEAL_PLAN_UTILIZATION+ PROMO_PARTICIPATION+
                                 CANCEL_RATE""", data = customer)


# telling Python to run the data through the blueprint
results_full = lm_full.fit()


# printing the results
print(results_full.summary())
```

```python
#x variables for normal y
x_var = ['TOTAL_MEALS_ORDERED','UNIQUE_MEALS_PURCH', 'PRODUCT_CATEGORIES_VIEWED','AVG_TIME_PER_SITE_VISIT',
        'MOBILE_LOGIN_RATIO','AVG_ORDER_VALUE','AVG_PREP_VID_TIME','AVG_MEAL_ORDER_PER_CUSTOMER',
        'AVG_MEAN_RATING','TOTAL_LOGIN','AVG_DELIVERY_TIME','CUSTOMER_ENGAGEMENT_SCORE',
        'MEAL_PLAN_UTILIZATION','CANCEL_RATE']

# x variables for log y
x_var_log_y = ['UNIQUE_MEALS_PURCH','CONTACTS_W_CUSTOMER_SERVICE','PRODUCT_CATEGORIES_VIEWED','AVG_TIME_PER_SITE_VISIT',
                'MEAL_PLAN_UTILIZATION','AVG_ORDER_VALUE','CUSTOMER_ENGAGEMENT_SCORE','AVG_PREP_VID_TIME',
                'AVG_MEAL_ORDER_PER_CUSTOMER','LOG_PROMO_PARTICIPATION','TOTAL_PHOTOS_VIEWED','CANCEL_RATE']
```

```python
# preparing for scikit-learn

# preparing x-variables from the OLS model
x_data     = customer[ x_var ]
log_x_data = customer[x_var_log_y]


# preparing response variable
y_data     = customer['REVENUE']
log_y_data = customer[ 'LOG_REVENUE' ]


###############################################
## setting up more than one train-test split ##
###############################################
x_train, x_test, y_train, y_test = train_test_split(
            x_data, # x-variables 
            y_data, # y-variable 
            test_size    = 0.25,
            random_state = 219)
```

```python
# Setting a model name
model_name = "Final Linear Regression Model (Using Scikit-Learn)"


# INSTANTIATING a model object - CHANGE THIS AS NEEDED
model = sklearn.linear_model.LinearRegression()


# FITTING to the training data
model_fit = model.fit(x_train, y_train)


# PREDICTING on new data
model_pred = model.predict(x_test)


# SCORING the results
model_train_score = model.score(x_train, y_train).round(4) # using R-square
model_test_score  = model.score(x_test, y_test).round(4)   # using R-square
model_gap         = abs(model_train_score - model_test_score).round(4)


# displaying results
print('Training Score :', model_train_score)
print('Testing Score  :', model_test_score)
print('Train-Test Gap :', model_gap)
```

```python
# zipping each feature name to its coefficient
model_coefficients = zip(x_train.columns,
                         model.coef_.round(decimals = 2))


# setting up a placeholder list to store model features
coefficient_lst = [('intercept', model.intercept_.round(decimals = 2))]


# printing out each feature-coefficient pair one by one
for coefficient in model_coefficients:
    coefficient_lst.append(coefficient)
    

# checking the results
for pair in coefficient_lst:
    print(pair)
```

```python
# dynamically printing results
ols_model =  f"""\
Model Name:     {model_name}
Train_Score:    {model_train_score}
Test_Score:     {model_test_score}
Train-Test Gap: {model_gap}

Coefficents
-----------
{pd.DataFrame(data = coefficient_lst, columns = ["Feature", "Coefficient"])}"""

print(ols_model) # printing the designed model
```
